```python
import xml.etree.ElementTree as ET
from nltk.tokenize import RegexpTokenizer
from nltk.stem import PorterStemmer
import math
import scipy
from operator import itemgetter
```

```python
def tokenizer(str1):
    str1=str1.lower()
    return str1.split(" ")
     
```

```python
stopwords = open('stoplist.txt','r')
stops = stopwords.readlines()
stoplist = []
for word in stops:
    stoplist.append(word.strip())
```

```python
def removeWords(listt, obj):
    for word in list(obj):
        
        if word in listt:
            obj.remove(word)
    return obj
```

```python
def Stemming(tokens):
    
    output=[]
    st = PorterStemmer()
    for line in tokens:
        output.append(" ".join([st.stem(i) for i in line.split()]))
    return output
```

```python
# Split the query into tokens (it is most correct to use the regular expression, but for these
# queries it suffices to split on whitespace).\\\\
#  Convert all tokens to lowercase.\\\\
#  Apply stop-wording to the query using the same list you used in assignment 1.\\\
#  Apply the same stemming algorithm to the query which you used in your indexer.\\\\

queryvecs={}

tree = ET.parse('topics.xml')
for node in tree.iter('topic'):
    query=node.find('query')
    #print(query.text)
    #print(node.attrib['number'])
    
    tokens=tokenizer(str(query.text))
    tokens1=removeWords(stoplist,tokens)
    tokens2=Stemming(tokens1)
    queryvecs[int(node.attrib['number'])]=tokens2

print(queryvecs)
```
